```python
import pandas as pd
import numpy as np
```

```python
angry = pd.read_excel("angry.xlsx")
happy = pd.read_excel("happy.xlsx")
love = pd.read_excel("love.xlsx")
sad = pd.read_excel("sad.xlsx")
depressed = pd.read_excel("depressed.xlsx")
```

```python
angry.columns=['sentence','label']
happy.columns=['sentence','label']
love.columns=['sentence','label']
sad.columns=['sentence','label']
depressed.columns=['sentence','label']
```

```python
angry = angry.dropna()
happy = happy.dropna()
love = love.dropna()
sad = sad.dropna()
depressed = depressed.dropna()
```

```python
data=pd.concat([angry,happy,love,sad,depressed])
```

```python
data = data.reset_index(drop=True)
```
