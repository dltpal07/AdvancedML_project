## querys와 label 변경하여 사용하기
## 43번째 줄 ws['A'+str(cnt)] = string 부분이 A열의 cnt 번째줄에 입력하는 거라서 cnt 겹치면 사라짐 그래서 만약에 슬픔 끝내고 설렘 돌릴 때는 cnt가 1이 아니라 최종 cnt부터 시작해야 함(중요 중요ㅠㅠㅠ)

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
from openpyxl import Workbook


# 생각해둔 키워드
querys = ['슬퍼', '속상해']
# 라벨 : 기쁨(0) 슬픔(1) 우울(2) 설렘(3) 화남(4) 상쾌(5)
label = 1



# 전체 샘플 문장 개수 카운트
cnt = 1
# 다음 문장과 같은자를 비교하기 위한 이전 문장
before = ""
# 페이지가 끝났을 경우를 확인하기 위한 카운트
# 이전 문장과 두 번 이상 비슷할 경우 검색어에 의한 페이지가 끝났다고 판단
mini_cnt = 0

wb = Workbook()
ws = wb.active
# 검색 페이지 수
page = 1

for q in querys:
    # 1 페이지부터 시작
    page = 1
    while(1):
        mini_cnt = 0
        req = requests.get("https://kin.naver.com/search/list.nhn?query=" + q +"&page="+str(page))
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        mytitles = soup.select(
            '#s_content > div.section > ul > li:nth-child(10) > dl > dt > a'
        )
        for i in range(1,11):
            # 문장 추출
            string = (soup.select('#s_content > div.section > ul > li:nth-child(%i) > dl > dt > a' %i))[0].text
            print(i, string)
            # 문장 내용
            ws['A'+str(cnt)] = string
            # 라벨
            ws['B'+str(cnt)] = label
            cnt = cnt+1
            if(before == string):
                mini_cnt = mini_cnt + 1
            else:
                mini_cnt = 0
            before = string
        wb.save("result.xlsx")
        page = page + 1
        if(mini_cnt > 1):
            mini_cnt = 0
            break
print(cnt)
```
